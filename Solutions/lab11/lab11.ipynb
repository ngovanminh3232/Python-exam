# Chuẩn Bị Dữ Liệu

```python
import pandas as pd
```

```python
df = pd.read_excel('provinces.xls')
df
```

# Bài 1. Mật Độ Dân Số

**Tính mật độ dân số**

```python
df['PopulationDensity'] = df['Population']*1000 / df['Area']
df
```

**Tìm tỉnh/thành phố có mật độ dân số cao nhất và thấp nhất**

```python
# max and min population density
max_dense = df['PopulationDensity'].max()
min_dense = df['PopulationDensity'].min()

# province/city with max and min population density
max_dense_province = df[df['PopulationDensity'] == max_dense].iloc[0, :]
min_dense_province = df[df['PopulationDensity'] == min_dense].iloc[0, :]

# print result
print('Most densely populated province/city: ' +
      f'{max_dense_province["Name"]}, {int(max_dense_province["PopulationDensity"])} people/km2')
print('Least densely populated province/city: ' +
      f'{min_dense_province["Name"]}, {int(min_dense_province["PopulationDensity"])} people/km2')
```

# Bài 2. Bờ Biển

**Đọc dữ liệu, tìm số tỉnh thành giáp biển**

```python
coast_df = pd.read_excel('coastlines.xls')

print('Number of provinces & cities with coastline:', len(coast_df))
coast_df.head()
```

**Nhập dữ liệu bờ biển vào dữ liệu chung các tỉnh thành**

```python
# initilize list to store coastline of each province
coastlines = []

# loop over provinces in `df`
for name in df['Name']:
    matching_row = coast_df[coast_df['Name'] == name]  # find matching row in `coast_df`
    if len(matching_row) == 0:  # matching row not found => no coastline
        coastlines.append(0)
    else:                       # matching row found => record to `coastlines`
        coastlines.append(matching_row['Coastline'].iat[0])

# add 'Coastline' column to `df` and save to excel file
df['Coastline'] = coastlines
df.to_excel('provinces_updated.xls', index=False)
df
```

# Bài 3. Biên Giới

**Đọc dữ liệu và nhập vào cùng một DataFrame**

```python
# initilize empty result DataFrame
df = pd.DataFrame(columns=['Name', 'BorderWith', 'BorderCount'])

# loop over countries, corresponding to each datasheet
for country in (('CHN', 'China'), ('LAO', 'Laos'), ('KHM', 'Cambodia')):
    
    # read data from a sheet
    inp_df = pd.read_excel('borders.xls', sheet_name=country[0])

    # loop over provinces in `inp_df`
    for name in inp_df['Tên tỉnh']:
        matching_row = df[df['Name'] == name]  # find matching row in `df`
        if len(matching_row) == 0:  # matching row not found => append new province to `df`
            df = df.append({'Name': name, 'BorderWith': country[1], 'BorderCount': 1}, ignore_index=True)
        else:                       # matching row found => update the row in `df`
            df.at[matching_row.index[0], 'BorderWith'] += ', ' + country[1]
            df.at[matching_row.index[0], 'BorderCount'] += 1

# save resulting DataFrame to file
df.to_excel('borders_updated.xls', index=False)
df
```
