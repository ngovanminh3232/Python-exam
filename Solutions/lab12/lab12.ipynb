```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Bài 1. Dân Số

```python
df = pd.read_excel('provinces_updated.xls')
df = df.sort_values('Population')
df
```

```python
# initialize figure, set size in inches
fig, ax = plt.subplots(figsize=(16, 6))

# bar plot: Population in Provinces
plt.bar(range(len(df)), df['Population'], width=0.6, tick_label=range(1, len(df)+1))

# customize axes: limit & visibility
plt.xlim(-1, 63)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# set title & axis labels
plt.suptitle('Population in Vietnam Provinces & Cities', fontsize=24, color='#444444', x=0.125, ha='left')
plt.title('sorted ascending', fontsize=18, color='#777777', x=0, ha='left')
plt.xlabel('Province / City', fontsize=18, color='#333333')
plt.ylabel('Population (1000 people)', fontsize=18, color='#333333')
plt.show()
```

**Chú thích:**

```python
print('Index of provinces & cities: ', end='')
for i, name in enumerate(df['Name']):
    print(f'{i+1}. {name}  ', end='')
```

# Bài 2. Diện Tích Vùng

```python
df = pd.read_excel('region_areas.xls')
df
```

```python
# initialize figure, set size in inches
fig, ax = plt.subplots(figsize=(16, 6))

# pie chart: proportion of area in regions
plt.pie(df['Area Proportion (%)'], autopct='%1.1f%%', startangle=90, counterclock=False,
        wedgeprops={'edgecolor': 'w', 'linewidth': 1}, labels=df['Region'])

# set title
plt.title('Proportion of Area in Regions', fontsize=20, color='#333333')

plt.show()
```

# Bài 3. Mật Độ Dân Số

```python
df = pd.read_excel('provinces_updated.xls')
df
```

```python
# initialize figure, set size in inches
fig, ax = plt.subplots(figsize=(16, 6))

# histogram: frequency of temperatures
n, bins, edges = plt.hist(df['PopulationDensity'], bins=20, ec='white')
plt.xticks(bins)

# customize axes: limit & visibility
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# set title, axis labels & legends
plt.title('Frequency of Population Density in Vietnam Provinces & Cities', fontsize=24, color='#444444', x=1, ha='right')
plt.xlabel('Population Density (people / km2)', fontsize=18, color='#333333')
plt.ylabel('Frequency', fontsize=18, color='#333333')

plt.show()
```

# Bài 4. Lốc Xoáy ở Việt Nam

```python
df = pd.read_csv('cyclones.csv')
df
```

```python
# initialize figure, set size in inches
fig, ax = plt.subplots(figsize=(16, 6))

# line plot: cyclone events over years
plt.plot(df['date'], df['cyclone_events'])

# customize axes: limit & visibility
plt.ylim(0, None)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# set title & axis labels
plt.title('Cyclone Events in Vietnam', fontsize=24, color='#444444', x=0, ha='left')
plt.xlabel('Year', fontsize=18, color='#333333')
plt.ylabel('Number of Events', fontsize=18, color='#333333')

plt.show()
```
